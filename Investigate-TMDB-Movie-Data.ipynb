### Research Question 1 (Which genres are receiving higher user ratings?)

```python
# top 10 genres with higher average user rating
top_10_genres = movie_df.groupby(['genres']).mean()['vote_average'].sort_values(ascending = False)[: 10]
top_10_genres
```

>### Action 1 for Q1
- For answering, which genres are receiving higher user ratings? I have used `groupby()` function to group by `genres` and then taking `mean()` found the average user ratings for top `10` genres.
- The below bar chart showcases the same thing visually.

```python
# function for plotting horizontal bar chart
def plot_barh(arr1, arr2, xlabel, ylabel, title):
    plt.barh(top_10_genres.index, top_10_genres.values)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.show()
```

```python
# plot for top 10 genres by user ratings
arr1 = top_10_genres.index
arr2 = top_10_genres.values
xlabel = 'Mean Vote Average'
ylabel = 'Genres'
title = 'Top 10 Genres by User Ratings'
plot_barh(arr1, arr2, xlabel, ylabel, title)
```

```python
# top 10 genres based on average vote count 
top_10_genres_count = movie_df.groupby(['genres']).mean()['vote_count'].sort_values(ascending = False)[: 10]
top_10_genres_count
```

>### Action 2 for Q1
- Same question is answered in different angle by using `vote_count` instead of `vote_average`.
- This time the top 10 genres are different from what I got in previous step.
- The below bar chart shows the top 10 genres based on average `vote_count`

```python
# plot for genres by average vote count
arr1 = top_10_genres_count.index
arr2 = top_10_genres_count.values
xlabel = 'Average Vote Count'
ylabel = 'Genres'
title = 'Top 10 Genres by Average Vote Count'
plot_barh(arr1, arr2, xlabel, ylabel, title)
```

### Research Question 4 (How is the revenue and popularity trend for movies from year to year?)

```python
# Calculate revenue of movies from year to year
revenue_by_years = movie_df.groupby(['release_year']).sum()['revenue_adj']
revenue_by_years.head()
```

>### Action 1 for Q4
- In order this question I have used `groupby()` function to group by years and then calculating the total revenue generated in each year.
- `revenue_by_years` is pandas series holding revenue generated in each year.
- Below plot shows the trend of revenue from year to year.

```python
# plot for revenue generated from year to year
plt.bar(revenue_by_years.index, revenue_by_years.values)
plt.plot(revenue_by_years, color = 'r', alpha = 0.5)
plt.xlabel('Year')
plt.ylabel('Revenue')
plt.title('Revenue Generated from Year to Year')
plt.show()
```

>### Conclusion 1 for Q4
- Based on the above visual for `Revenue Generated From Year to Year` it is found that revenue generated by movies has increased as years progressed which means more people are watching movies.
- Yes, there are some years where there is drop in revenue but if we consider as a whole the trend of the revenue seems like increasing.
- One important point we have to consider here is to observe the trend of number of movies released from year to year. So, let's see the trend.

```python
# calculate number of movies from year to year
n_movies_by_years = movie_df.groupby(['release_year']).count()['id']
n_movies_by_years.head()
```

>### Action 2 for Q4
- In order to see the number of movies released from year to year, I have used `groupby()` function to group by `release_year` and then count to find the number of movies in each year
- Below plot shows the trend of number of movies released from year to year.

```python
# plot for number of movies from year to year
plt.bar(n_movies_by_years.index, n_movies_by_years.values)
plt.plot(n_movies_by_years, color = 'r', alpha = 0.5)
plt.xlabel('Year')
plt.ylabel('Number of Movies')
plt.title('Number of Movies Released from Year to Year')
plt.show()
```

>### Conclusion 2 for Q4
- Based on the above visual for `Number of Movies Released from Year to Year` it is found that number of movies released has increased as years progressed. 
- Which implies that revenue generated will also be increased from year to year.

```python
# calculate popularity of movies from year to year
popularity_by_years = movie_df.groupby(['release_year']).mean()['popularity']
popularity_by_years.head()
```

>### Action 3 for Q4
- In order to observe the trend of popularity of movies from year to year, I have used `groupby()` function to group by `release_year` and then taking mean popularity of movies in each year.
- Below plot shows the trend of popularity of movies from year to year.

```python
# plot popularity of movies from year to year
plt.bar(popularity_by_years.index, popularity_by_years.values)
plt.plot(popularity_by_years, color = 'r', alpha = 0.5)
plt.xlabel('Year')
plt.ylabel('Average Popularity of Movies')
plt.title('Popularity of Movies from Year to Year')
plt.show()
```

>### Conclusion 3 for Q4
- Based on the above visual for `Popularity of Movies from Year to Year` it is found that popularity of movies has increased as years progressed but the increase is not very steep until `2011`, after `2011` the increase in popularity is very steep.
- This implies popularity for movies has increased very much in `20's`
